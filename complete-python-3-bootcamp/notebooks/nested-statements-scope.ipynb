# Nested Statements and Scope in Python

This notebook explains two key concepts:

1. Nested statements
2. Variable scope (LEGB rule)

Understanding scope helps avoid common errors like `NameError: name 'x' is not defined`.

## 1) Namespaces and Variable Scope

Python stores names in namespaces. Scope determines where a name can be accessed.

If you try to access a variable outside its scope, Python raises an error.

```python
def my_function():
    local_var = "I live inside the function"
    print(local_var)

my_function()

# Uncommenting the next line would raise NameError
# print(local_var)
```

## 2) The LEGB Rule

Python resolves variable names in this order:

- **L**ocal: inside the current function
- **E**nclosing: inside outer functions (nested functions)
- **G**lobal: top-level module variables
- **B**uilt-in: Python reserved names like `len`, `sum`, etc.

```python
x = "global x"

def outer():
    x = "enclosing x"

    def inner():
        x = "local x"
        print("Inside inner():", x)

    inner()
    print("Inside outer():", x)

outer()
print("At global level:", x)
```

## 3) Assignment Behavior Across Scopes

Assigning a variable inside a function creates a local variable by default.

That means reassigning `x` inside a function does **not** automatically modify global `x`.

```python
x = 25

def tester():
    x = 50  # local assignment
    return x

print("Result from tester():", tester())
print("Global x after tester():", x)
```

## 4) Using the `global` Keyword

`global` lets you modify a global variable inside a function.

Use this carefully, because it can make code harder to reason about.

In many cases, passing values as function parameters is cleaner and safer.

```python
x = 10

def modify_global():
    global x
    x = 99

print("Before modify_global():", x)
modify_global()
print("After modify_global():", x)
```

## 5) Safer Alternative to `global`

A cleaner approach is to pass values in and return results out.

```python
def increment(value):
    return value + 1

x = 10
x = increment(x)
print("Updated x:", x)
```

## Summary

- Scope controls where variables are visible and usable.
- Python searches names using LEGB: Local, Enclosing, Global, Built-in.
- Assignments inside functions are local unless `global` is used.
- Prefer parameter passing and return values for cleaner code.
